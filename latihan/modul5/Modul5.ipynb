```python
import matplotlib.pyplot as plt
from numpy.random import normal
from numpy import std, mean

sample = normal(loc=50, scale=5, size=1000)

plt.figure(figsize=(5,4))
plt.hist(sample, bins=10, density=True)
plt.show()

sample_mean=mean(sample)
sample_std=std(sample)

print('Mean=%.3f \nStandard Deviation=%.3f' % (sample_mean, sample_std))

dist = norm(sample_mean, sample_std)
print(dist)

values = [value for value in range(30,70)]
probabilitas = [dist.pdf(value) for value in values]
probabilitas

plt.hist(sample, bins=10, density=True)
plt.plot(values, probabilitas)
plt.show()
```

```python
#KEGIATAN 1

import matplotlib.pyplot as plt
from functools import reduce

nilai_mahasiswa = [75, 80, 90, 65, 70, 85, 95, 78, 88, 92]

def rata_rata(data):
 total = reduce(lambda x, y : x+y, data)
 return total / len(data)

hasil_rata_rata = rata_rata(nilai_mahasiswa)

mahasiswa = list(map(lambda x: f"Mahasiswa {x + 1}", range(len(nilai_mahasiswa))))

plt.bar(mahasiswa, nilai_mahasiswa, color='blue')
plt.xlabel('Mahasiswa')
plt.ylabel('Nilai Ujian')
plt.title('Diagram Batang Nilai Ujian Mahasiswa')

plt.axhline(y=hasil_rata_rata,  color='red', linestyle='--', label=f'Rata-rata: {hasil_rata_rata:.2f}')

for i, value in enumerate(nilai_mahasiswa):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm  # Import modul norm dari scipy.stats

tinggi_badan = [165, 170, 155, 172, 180, 160, 175, 165, 185, 175, 170, 160]
interval_size = 10
bins = [150, 155, 160, 165, 170, 175, 180, 185, 190]

def buat_histogram(data, bins, interval_size):
    plt.hist(data, bins=bins, density=True, label='Histogram')  # Menggunakan density=True
    mu, std = np.mean(data), np.std(data)
    x = np.linspace(min(data), max(data), 100)
    pdf = norm.pdf(x, mu, std)
    plt.plot(x, pdf, color='red', label='PDF')
    plt.xlabel('Tinggi Badan')
    plt.ylabel('Frekuensi / PDF')
    plt.title('Histogram dan Kurva PDF Tinggi Badan')
    plt.legend()
    plt.show()

def kelompok_data(data, bins, interval_size):
    grouped_data = np.histogram(data, bins=bins)
    print("Interval: ", grouped_data[1])
    print("Frekuensi: ", grouped_data[0])


kelompok_data(tinggi_badan, bins, interval_size)
```
